```python
from geopy.distance import geodesic
```

```python
# Function to calculate distance between two sets of coordinates
def calculate_distance(coord1, coord2):
    return geodesic(coord1, coord2).kilometers
```

```python
# Convert latitude and longitude strings to floats and calculate distance
for record in records:
    latitude, longitude = record['New Georeferenced Column'].split("(")[-1].split(")")[0].split()
    record['longitude'] = float(longitude)
    record['latitude'] = float(latitude)
```

```python
# Calculate distance between each pair of coordinates
for i in range(len(records)):
    for j in range(i+1, len(records)):
        coord1 = (records[i]['latitude'], records[i]['longitude'])
        coord2 = (records[j]['latitude'], records[j]['longitude'])
        distance = calculate_distance(coord1, coord2)
        print(f"Distance between {records[i]['Station Name']} and {records[j]['Station Name']}: {distance:.2f} km")
    
```

```python
# Function to calculate distance between the center and a set of coordinates
def calculate_distance_to_center(coord, center_coord):
    return geodesic(coord, center_coord).kilometers
# Create a map
map_center = [53.06393526014041, -3.0084894057980054]
map = folium.Map(location=map_center, zoom_start=10.5)

# Add markers and lines to the map
for record in records:
    # Extract latitude and longitude from the record
    latitude, longitude = record['New Georeferenced Column'].split("(")[-1].split(")")[0].split()
    latitude, longitude = float(latitude), float(longitude)

    # Calculate distance from the center to the current coordinates
    distance_to_center = calculate_distance_to_center((latitude, longitude), map_center)

    # Add marker
    coords = (latitude, longitude)
    folium.Marker(coords, popup=f"Distance: {distance_to_center:.2f} km").add_to(map)

    # Add line from the center to the current coordinates
    folium.PolyLine([map_center, coords], color='blue').add_to(map)

# Display the map
map
```
